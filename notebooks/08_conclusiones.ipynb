# 5. Conclusiones

Resumen final del proyecto CRISP-DM sobre videos en tendencia de YouTube en Gran Bretaña. Se organiza en las cuatro subsecciones exigidas por la rubrica del trabajo.

## 5.1 Conclusiones de negocio

- **Music** y **Entertainment** son las categorias con mayor volumen de apariciones en tendencia, lo que indica que el consumo de contenido en YouTube GB se concentra fuertemente en ocio.
- Las categorias con **mas likes** promedio no siempre coinciden con las de mayor volumen; existen nichos (p. ej. Sports, Science) con alta aceptacion relativa aunque menor presencia.
- El **ratio Likes/Dislikes** y el **ratio Vistas/Comentarios** permiten priorizar categorias por calidad de recepcion, no solo por cantidad de videos.
- Un **grupo reducido de canales** concentra la mayoria de apariciones en tendencia (ley de Pareto), lo que sugiere que la plataforma premia la consistencia y autoridad del canal.
- La prediccion de **vistas** es factible con variables de interaccion y temporales, aunque el error sigue siendo alto debido a la naturaleza viral de los contenidos.

### Implicaciones para el cliente de marketing digital
1. **Apostar por categorias de alto engagement** como Music y Entertainment, pero tambien evaluar nichos bien valorados.
2. **Monitorear ratios de interaccion** como KPI de calidad de contenido, no solo volumen de vistas.
3. **Colaborar con canales recurrentes** en tendencia para aprovechar su traccion organica.
4. **Publicar y ajustar horarios** segun los patrones temporales detectados en la fase de exploracion.

## 5.2 Conclusiones tecnicas

- **Procesamiento**: la limpieza fue ligera; la mayoria de outliers representan viralidad real y no fueron eliminados. Se filtraron inconsistencias logicas (interacciones mayores que vistas, geografia invalida) y categorias no mapeadas.
- **Feature engineering**: las variables derivadas (`engagement_rate`, `likes_per_view`, `days_to_trending`, `views_per_comment`) aportaron mayor senal predictiva que las metricas absolutas.
- **Modelado**: Random Forest y Gradient Boosting superaron al modelo lineal y al baseline. El uso de `log1p(views)` estabilizo la varianza y mejoro las metricas.
- **Evaluacion**: el analisis de residuos mostro error heterocedastico creciente con el volumen, coherente con la ley de Pareto. El diagnostico de sesgo/varianza indico un balance aceptable, aunque con margen de mejora.
- **Cobertura de requerimientos**: los 9 requerimientos del cliente se responden con visualizacion y tabla; la verificacion se automatiza en `09_verificacion_cumplimiento.ipynb`.

## 5.3 Limitaciones

- **Geografia simulada**: las columnas `state`, `lat` y `lon` fueron asignadas de forma aleatoria por el dataset academico; los hallazgos geograficos son ilustrativos.
- **Sin texto de comentarios**: el dataset no incluye los comentarios de los usuarios, por lo que el analisis de sentimiento (R8) usa el **titulo** como proxy.
- **Ventana temporal limitada**: el dataset cubre un periodo corto; no se pueden inferir tendencias estacionales de largo plazo.
- **Sesgo de popularidad**: los datos solo incluyen videos que ya llegaron a tendencias, por lo que el modelo no predice la probabilidad de llegar a trending desde cero.
- **Colinealidad entre interacciones**: likes, dislikes y comentarios estan correlacionados con vistas, lo que dificulta identificar causalidad.
