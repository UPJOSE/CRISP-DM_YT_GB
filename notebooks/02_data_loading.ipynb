# 2. Comprensión de los Datos — Carga y Diccionario

```python
import json
import pandas as pd
import numpy as np

# ==========================================
# Rutas del dataset (carpeta data del proyecto)
# ==========================================
DATA_DIR = "../data"
CSV_PATH = f"{DATA_DIR}/GBvideos_cc50_202101.csv"
JSON_PATH = f"{DATA_DIR}/GB_category_id.json"
```

## 2.1 Carga de archivos

```python
# Carga del CSV principal
df = pd.read_csv(CSV_PATH)

# Carga del JSON de categorías y mapeo category_id -> nombre
with open(JSON_PATH, "r", encoding="utf-8") as f:
    categories_data = json.load(f)

category_map = {
    int(item["id"]): item["snippet"]["title"] for item in categories_data["items"]
}
df["category_name"] = df["category_id"].map(category_map).fillna("Unknown")

print("Shape:", df.shape)
print("Columnas:", df.columns.tolist())
```

## 2.3 Diccionario de datos

```python
# Diccionario de datos: tipo, nulos y descripción de cada variable
descripciones = {
    "video_id": "Identificador único del video",
    "title": "Título del video",
    "channel_title": "Nombre del canal",
    "category_id": "ID numérico de categoría",
    "category_name": "Nombre descriptivo de la categoría",
    "tags": "Etiquetas separadas por '|'",
    "description": "Descripción textual del video",
    "thumbnail_link": "URL de la miniatura",
    "publish_time": "Fecha y hora de publicación",
    "trending_date": "Fecha en la que el video apareció en tendencias",
    "views": "Número de vistas (variable objetivo)",
    "likes": "Número de likes",
    "dislikes": "Número de dislikes",
    "comment_count": "Número de comentarios",
    "comments_disabled": "Comentarios desactivados",
    "ratings_disabled": "Calificaciones desactivadas",
    "video_error_or_removed": "Video con error o eliminado",
    "state": "Región del Reino Unido",
    "lat": "Latitud geográfica",
    "lon": "Longitud geográfica",
    "geometry": "Geometría WKT (POINT lon lat)",
}

dict_df = pd.DataFrame(
    {
        "variable": df.columns,
        "tipo": [str(df[c].dtype) for c in df.columns],
        "nulos": [df[c].isna().sum() for c in df.columns],
        "pct_nulos": [round(df[c].isna().mean() * 100, 2) for c in df.columns],
        "descripcion": [descripciones.get(c, "Pendiente de documentar") for c in df.columns],
    }
)
display(dict_df)
```
